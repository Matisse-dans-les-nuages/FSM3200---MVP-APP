```python
#Fonction de numérisation des labels, pour les datafrane

def hm_to_label(hm):
    hm_list = ['Small particle',
                   'Columnar crystal',
                   'Planar crystal',
                   'Columnar/Planar crystals',
                   'Aggregate',
                   'Graupel',
                   'None']
    try:
        #On donne le label propre à une catégorie
        manual_label=hm_list.index(hm)+1
    except:
        #Dans les instances NaN, on veut le label None (7)
        manual_label=7
    return manual_label
```

```python
# Fonction de filtre des images corrompues. Renvoie un dataframe sans les images corrompues, 
# et seulement avec les images classées (Save=1)
def complete_filter(df):
    return df[(df['radio variable'] == 0) &
              (df['saves'] == 1) &
              (df['manual ID'].isin([1,2,3,4,5,6]))]
```

```python
def calculate_invalid_rate(df):
    """
    Calcule le taux d'images invalides dans un DataFrame.
    Une image est considérée invalide si :
    - 'saves' est égal à 1, ET
    - 'manual ID' n'est pas compris entre 1 et 6 (inclus), OU
    - 'radio variable' est égal à 1.
    
    :param df: DataFrame Pandas contenant les colonnes 'saves', 'manual ID' et 'radio variable'.
    :return: Le taux d'images invalides (float).
    """
    # Condition pour 'saves == 1'
    saves_condition = df['saves'] == 1
    
    # Condition pour 'manual ID' non valide
    invalid_manual_id = ~df['manual ID'].between(1, 6)
    
    # Condition pour 'radio variable == 1'
    invalid_radio_variable = df['radio variable'] == 1
    
    # Combiner les conditions (appliquer les autres conditions seulement si 'saves == 1')
    invalid_condition = saves_condition & (invalid_manual_id | invalid_radio_variable)
    
    # Nombre total d'images où 'saves == 1'
    total_saves = saves_condition.sum()
    
    # Nombre d'images invalides
    invalid_images = invalid_condition.sum()
    
    # Calcul du taux d'images invalides
    invalid_rate = invalid_images / total_saves if total_saves > 0 else 0
    
    return invalid_rate

```

```python
for i in range(len(tempetes)):
    print(f"Tempête {i}, {calculate_invalid_rate(tempetes[i])}")
```

```python
# Matrices de confusion pour les classements complétés
def confusion_matrix(df, col1, col2):
    """
    Crée une matrice de confusion à partir de deux colonnes d'un DataFrame.
    
    :param df: Le DataFrame contenant les données.
    :param col1: Nom de la colonne pour les valeurs réelles.
    :param col2: Nom de la colonne pour les prédictions.
    :return: Une matrice de confusion sous forme de DataFrame.
    """
    # Générer la matrice de confusion
    confusion_matrix = pd.crosstab(df["manual ID"], df["label ID "], rownames=['Actual'], colnames=['Predicted'], dropna=False)
    
    # Ajouter des valeurs manquantes pour s'assurer que toutes les valeurs de 1 à 6 sont incluses
    index = range(1, 7)  # Valeurs possibles
    confusion_matrix = confusion_matrix.reindex(index=index, columns=index, fill_value=0)
    
    return confusion_matrix

```
