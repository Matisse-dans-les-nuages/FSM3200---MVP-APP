# Fonctions

```python
#Remplacer ".mat" par ".png" dans le dataframe contenant les noms de fichiers

def mat_to_png(filename):
    return filename.replace(".mat",".png")
```

```python
#Obtenir un datetime en fonction du nom de fichier

def filename_to_datetime(filename):
    #Assume format AAAA.MM.DD_HH.MM.SS_flake_x_cam_y.png
    #Change filename to datetime
    time = filename.split("_")[0]+"_"+filename.split("_")[1]
    timeformat = '%Y.%m.%d_%H.%M.%S'
    dt_object = datetime.strptime(time, timeformat)
    return dt_object
```

```python
#Obtenir un chemin relatif en fonction du nom de fichier

def path_from_filename(filename):
    path= "" # TODO
    filename = filename.replace('.png ','')
    timestamp = filename[0:18]
    month_file = timestamp[0:4]+"_"+timestamp[5:7]+"/"
    day_file = ".".join(timestamp[0:10].split("-"))+"/"
    hour = timestamp[11:13]+"/"
    
    return path+month_file+day_file+hour+filename+'.png'
```

## DATAFRAME POUR CHAQUE TEMPÊTE DE NEIGE

### FONCTIONS

```python
#Extraire en tuple le timestamp de début et de fin d'une tempête à partir d'un df donné
def timeframe(storm,storm_df):
    start=pd.to_datetime(storm_df['start'][storm])
    end=pd.to_datetime(storm_df['end'][storm])
    return (start,end)
```

```python
#Générer une section de dataframe en fonction de l'intervalle de la tempête choisie
def dataframe_from_storm_timeframe(storm,masc_df):
    start = storm[0]
    end = storm[1]
    storm_mdf = masc_df[(masc_df.index >= start) & (masc_df.index <= end)]
    return storm_mdf.copy()
```

```python

def identify_month(dt):
    """
    Identifies the month from a datetime object.
    
    Parameters:
    dt (string): A timestamp
    
    Returns:
    str: The name of the month
    """
    month_list = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

    month = month_list[ int(dt.split('-')[1]) -1]
    
    
    return month
```

### Création des dataframes propres à chaque tempête

```python
#---
# IMPORTER LA LISTE DES TEMPÊTES
#---
path_stormDates = "../../codes_export/dataframe/ETE24/Snowstorms_For_MASC_Validation.csv"
stormDates_df = pd.read_csv(path_stormDates)
stormDates_df.index.name = 'storm number'
stormDates_df=stormDates_df.drop("Unnamed: 0", axis=1)

print(stormDates_df)
```

```python
# Créer les dataframes propres à chaque tempête
for i in stormDates_df.index:
    stormTimeframe = timeframe(i, stormDates_df)
    month = identify_month(str(stormTimeframe[0]))
    
    if month == 'March':
        temporary_dataframe = dataframe_from_storm_timeframe( stormTimeframe,filtered_march_df )
        df_filename = f"march_storm0{i}_df.csv"
        temporary_dataframe.to_csv("../../codes_export/dataframe/ETE24/phase1_SN/MVP_setup_dataframes/march/"+"filtered_"+df_filename)
        
    elif month == 'April':
        temporary_dataframe = dataframe_from_storm_timeframe( stormTimeframe,filtered_april_df )
        df_filename = f"april_storm0{i}_df.csv"
        temporary_dataframe.to_csv("../../codes_export/dataframe/ETE24/phase1_SN/MVP_setup_dataframes/april/"+"filtered_"+df_filename)
        
    
    
```
